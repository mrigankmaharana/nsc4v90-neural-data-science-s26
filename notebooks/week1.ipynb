# 1. How to create variables


Python allows us to define variables and manipulate them. Variables can store numbers or strings. A number can be an **integer** (i.e. numbers such as 1, 5, 813 etc.) or a **float** (i.e. numbers such as 1.2, 3.5, 100.05). **Strings** are sequences of characters inside single or double quotation marks.

```python
x = 1
y = 2.0
z = 'Hello world!'
```

The `print` function allows us to display things such as the value of the variable.

```python
x = 1
y = 2.0
z = 'Hello world!'
print(x)
print(y)
print(z)
```

You can also display properties of the variable (e.g., what type of variable is it? integer, float, string etc.).

```python
x = 1
y = 2.0
z = 'Hello world!'
print(type(x))
print(type(y))
print(type(z))
```

Variable names may be alphanumeric (e.g. `x1`), but should not begin with a number (so don't use `1x` as a variable name). Variable names may also contain underscores (e.g. `x_1`, `_x1`) but no other special characters.

Basic arithmetic operations work as you would expect.

```python
print(x + y)
print(3*y)
print(y**3)
```

The output of an operation can be stored in a new variable if you want.

```python
z = x + y
print(z)
```

We can also reassign i.e., overwrite a variable.

```python
y = 4.0*y  # What the hell?!
print(y)
```

Remember that it's often useful to add comments such as the one above to your code using # so that you'll know what you were thinking when you go back to read it later!

A **list** is just what it sounds like. Lists are enclosed by square brackets. It can be a list of numbers, a list of strings, or even a list of lists.

# 3. How to access a particular element

When you have a list of numbers or strings, it is often useful to access specific elements of the list. This is called `indexing`.

We can access the `i`th element of a list using `my_list[i]`. We refer to `i` as an **index**, which refers to an element of the list named `my_list`.

```python
print(list1[1])
```

Were you expecting this to return 1 since that's the first element of `list1`? Be careful, since indexing starts with 0 in Python. Here are all of the elements in the list:

```python
print(list1[0], list1[1], list1[2])
```

A handy trick is to count backwards from the end of a list using negative numbers.

```python
print(list1[-1], list1[-2], list1[-3])
```

We'll discuss more about indexing later on when we introduce arrays.

# 4. How to loop through elements

One reason that computers are so useful is because they are great at doing similar things many times over. The `for` loop is one of the basic structures that allows us to do this. Here is a very basic example:

```python
for n in [1,3,5]:
    print(2*n)
```

The above example is called a *for loop* because the indented part is repeated several times (i.e. looped), with `n` taking one of the values in the list during each iteration. We can also loop through a list that was already created instead of explicitly creating it as part of the *for loop*.

```python
list_of_numbers = [5, 12, 34, 90, 15]
for n in list_of_numbers:
    print(2*n)
```

A very handy thing to use with for loops is `range()`, which essentially creates a sequence with a range of sequential values (starting with 0, not 1!):

```python
for n in range(5): #Generates sequence of numbers from 0 to 4 doesn't include 5. Follows range(start, stop, step)
    print(n)
```

As another example, let's make a counter that counts how many times we have iterated over the loop.

```python
count = 0
for n in range(5):
    count += 1  # Note: This is a shortcut notation that does the same thing as count=count+1
    print(count)
```

We can include more than one operation inside of the for loop, remembering to indent everything that we want to be looped over.

```python
for n in range(5):
    n_squared = n**2
    n_cubed = n**3
    print(n_squared, n_cubed)

    #0, 1, 2, 3, 4
    #squared = 0, 1, 4, 9, 16
    #cubed = 0, 1, 8, 27, 64
```

Rather than printing each iteration, let's replace each element in a list with its square and then print it at the end. Note that, if a command appears with no indent, that signals that the loop has ended.

```python
n_list = [1,2,3,4,5]

for i in range(len(n_list)): #Loop over indices of the list, len() gives the length of the list which is 5, range of 5 is 0 to 4, for i in [0,1,2,3,4] we iteratre through n_list[0], n_list[1], n_list[2], n_list[3], n_list[4] and update those values by squaring them and then printing the new list.
    n = n_list[i] #Get the value at index i
    n_list[i] = n**2 #Update the value at index i to be n squared

print(n_list) #Should print [1, 4, 9, 16, 25]
```

Note that in the above example we had to use `i` to keep track of the index and `n` to keep track of the corresponding value in the array. A handy way to shorten things is to use `enumerate`.

```python
n_list = [1,2,3,4,5]

for i, n in enumerate(n_list):  # The first variable after 'for' is the index, and the second is the corresponding value, enumerate() generates pairs of (index, value) from the list.
    n_list[i] = n**2

print(n_list)
```

# 5. How to check if a criterion is met

We can check if a criterion is met using `conditional statements` . Such statements return either true or false. Some examples are "greater than", "less than" and "equal to", as shown in the following examples.

```python
print(1 > 2)
print(1 < 2)
```

```python
print(1 == 2)  # Note that testing equality requires two equal signs, unlike the assignment operator "=" we've seen so far.
print(2 == 2)
print(1 != 2)  # The operator "!=" means "not equal"
```

The output you see above i.e., True or False is a special variable called a Boolean. There are also operations that allow us to manipulate Boolean values using logical rules.

```python
# For AND, both things being compared must be True for the result to be True:
print(True and True)
print(True and False)
print(False and False)
```

```python
# For OR, at least one of the things must be True for the result to be True:
print(True or True)
print(True or False)
print(False or False or True)
print(False or False)
```

The control statements `if`, `elif`, and `else` enable us to choose what to do depending on Boolean values. As with for loops, all of the commands that are subject to the control statements must be indented. Below are a few examples.

```python
x = 0
if x < 10:
    print('x is less than 10')  # This part is run only if the control statement is True
```

```python
x = 12
if x < 10:
    print('x is less than 10')  # This part is run only if the control statement is True
```

```python
x = 12
if x < 10:
    print('x is less than 10')  # This part gets ignored since the control condition is False
elif x >= 10:  # "elif" means "else if", i.e. check this condition if the first condition was False
    print('x is greater than or equal to 10')
```

```python
x = 12
if x < 10:
    print('x is less than 10')
elif x == 10:
    print('x is equal to 10')
else:  # This part is run if everything tried above was False.
    print('x is greater than (and not equal to) 10')
```
